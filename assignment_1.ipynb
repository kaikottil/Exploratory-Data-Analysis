```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("./data/creditcard.csv")
data.head()
```

```python
shape = data.shape
print('\nDataFrame Shape :', shape)
```

```python
stats = data['Class'].value_counts()
print(stats)
```

```python
subset=data[['Class','Time','V1','V2']]
subset.head()
```

```python
subset_shape = subset.shape
print('\nDataFrame Shape with duplicates:', subset_shape)
subset = subset.drop_duplicates(keep='first')
subset_shape = subset.shape
print('\nDataFrame Shape without duplicates:', subset_shape)
```

```python
subset.groupby('Class').count()
```

```python
colors = ["#00FF00", "#FF0000"]
sns.displot(subset, x="Time", hue="Class", kind="kde", palette=colors)
```

```python
sns.pairplot(subset, hue="Class", palette=colors)
```
